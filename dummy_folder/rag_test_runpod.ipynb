### Hugging Face 테스트 (Naver Hyper Clova)

````python
from langchain.schema import Document

docs = []
for d in data:
    paragraphs = d["page_content"].split("\n\n")
    for p in paragraphs:
        if p.strip():  # 빈 문자열은 제외
            docs.append(
                Document(
                    page_content=p.strip(),
                    metadata=d["metadata"]
                )
            )
````

# RAG Workflow

````python
raw_query = "나 주행중인데 갑자기 핸들이 안 움직여"
````

````python
from langchain_core.language_models import LLM
from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import CrossEncoder
import torch

device = "cuda"

# 1. 모델 로드
model_name = "naver-hyperclovax/HyperCLOVAX-SEED-Text-Instruct-1.5B"
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# rerakner 모델
reranker_model = CrossEncoder("dragonkue/bge-reranker-v2-m3-ko", device=device)

# 2. Query rewriting
rewrite_prompt = [
    {"role": "system", "content": "너는 운전 용어 질의 변환기야."},
    {"role": "user", "content": f"""
    사용자 질문: {raw_query}

    raw_query를 교통안전 매뉴얼에서 사용할 만한 표준 쿼리로 변환해.
    
    <변환 예시>
    - 깜깜하다/너무 어둡다 -> 야간 운전 요령
    - (고라니/멧돼지/고양이/개/강아지 등) 동물을 쳤다 -> 로드킬 대처 방법
    - 비/눈가 너무 많이 온다 -> 눈길/빙판길/우천 시 안전 운전 요령
    - 갑자기 차가 속도를 높힌다/막 튀어나간다 -> 급발진 대처법
    - 브레이크가/핸들/엑셀 등이 말을 안 들어 -> 브레이크/핸들/엑셀 등 작동 불능 대처법

    출력은 변환된 쿼리 문자열 한 줄만 작성하라. 다른 설명이나 괄호는 쓰지 마라.
    """}
]

rewrite_inputs = tokenizer.apply_chat_template(
    rewrite_prompt, add_generation_prompt=True, return_dict=True, return_tensors="pt"
).to(device)

rewrite_ids = model.generate(
    **rewrite_inputs,
    max_length=512,
    stop_strings=["<|endofturn|>", "<|stop|>"],
    tokenizer = tokenizer
)

raw_input_length = rewrite_inputs["input_ids"].shape[1]
re_gen_tokens = rewrite_ids[0][raw_input_length:]
re_answer = tokenizer.decode(re_gen_tokens, skip_special_tokens=True).strip()
print("Rewritten query:", re_answer)
````

````python
# 3. Retriever 검색
retriever = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 5, "lambda_mult": 0.5}
)
retriever_results = retriever.invoke(re_answer)


# 4. Reranker 적용
pairs = [(re_answer, doc.page_content) for doc in retriever_results] # (query, 문서본문) 쌍 생성
scores = reranker_model.predict(pairs) # reranker 점수 계산

# 점수 높은 순서대로 정렬
scored_results = sorted(
    zip(retriever_results, scores),
    key=lambda x: x[1],
    reverse=True
)

# reranker 1등 문서 선택 → 카테고리 결정
best_doc, best_score = scored_results[0]
target_category = best_doc.metadata.get("상황카테고리", None)

# 같은 카테고리 문서 전체 모으기
category_docs = [doc for doc in docs if doc.metadata.get("상황카테고리") == target_category]
# 디버깅용 context
context_text_debug = "\n\n---\n\n".join(
    [f"[출처: {doc.metadata.get('source','?')}] {doc.page_content}" for doc in category_docs]
)

# 실제용 context 생성 (같은 카테고리 문서 텍스트를 하나로 합치기)
context_text = " ".join([doc.page_content for doc in category_docs])


## 4. 최종 답변 생성
chat = [
    {"role": "system", "content": "너는 초보운전자 어시스턴트야."},
    {
        "role": "user", "content": 
        f"""
        아래 raw_query, re_answer, context는 네가 내부적으로만 참고하는 자료다. 이 자료들을 그대로 출력하거나 언급하지 마라.
        
        - query: {raw_query}
        - re-written query: {re_answer}
        - context: {context_text}

        다음을 지시상항에 따라라서 답해라:
        1) 사용자 raw_query와 re-answer를 통해 상황의 응급/급박 여부를 내부적으로만 판단하라.
        2) 내부적으로 판단이 마친 상태에서 다음 중 하나의 상황만을 고려하여 답해라.
            2.1) 사용자가 응급/급박 상황이라면:
                - 지금 context에사 당장 실행해야 할 행동 지침만 추출하여 몇 단어로 짧게 제시하라.
            2.2) 사용자가 현재 응급/급박이 아니거나 벗어난 상황이라면:
                - context에서 응급 상황 이후의 대처법을 추출하여 전달하라.
        """
    }
]

# 프롬프트 전달 & 답변 생성
inputs = tokenizer.apply_chat_template(chat, add_generation_prompt=True, return_dict=True, return_tensors="pt")
inputs = inputs.to(device)
output_ids = model.generate(
    **inputs,
    max_length=1024,
    stop_strings=["<|endofturn|>", "<|stop|>"],
    tokenizer=tokenizer
    )

# 생선된 부분만 답변
input_length = inputs["input_ids"].shape[1]
gen_tokens = output_ids[0][input_length:]
answer = tokenizer.decode(gen_tokens, skip_special_tokens=True)
print(answer)
````
